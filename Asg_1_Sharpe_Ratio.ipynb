```python
import numpy as np
import pandas as pd
df=pd.read_csv('sample_prices.csv')
df['BLUE Returns']=0
df['ORANGE Returns']=0
for i in range(12):
    df.loc[i+1,'BLUE Returns']=(df.loc[i+1,'BLUE']-df.loc[i,'BLUE'])/df.loc[i,'BLUE']*100
for i in range(12):
    df.loc[i+1,'ORANGE Returns']=(df.loc[i+1,'ORANGE']-df.loc[i,'ORANGE'])/df.loc[i,'ORANGE']*100
x1=df['BLUE Returns'].tolist()
y1=np.array(x1)
std_dev_BLUE=np.std(y1)
x2=df['ORANGE Returns'].tolist()
y2=np.array(x2)
std_dev_ORANGE=np.std(y2)
df['BLUE Sharpe']=0
df['ORANGE Sharpe']=0
for i in range(13):
    df.loc[i,'BLUE Sharpe']=(np.average(x1)-3)/std_dev_BLUE
for i in range(13):
    df.loc[i,'ORANGE Sharpe']=(np.average(x2)-3)/std_dev_ORANGE
a1=1
for i in x1:
    a1=a1*(1+(i/100))
annual_return_BLUE=(a1-1)*100
a2=1
for i in x2:
    a2=a2*(1+(i/100))
annual_return_ORANGE=(a2-1)*100
print('Annual returns of Blue stock are: ',annual_return_BLUE,'percent')
print('Annual returns of Orange stock are: ',annual_return_ORANGE,'percent')
## NOTE: I have assumed the risk free rate of return to be 3%.
## NOTE: I have found the Sharpe Ratio for the average monthly return for both blue as well as orange stocks, 
##       however to maintain uniformity, I have printed the single entry 13 times, as a column.
```
